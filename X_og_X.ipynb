```python
import pandas as pd
import numpy as np

df = pd.read_excel('concordance_preloaded_notenten23_tt1_20250617162822.xlsx')
```

```python
df
```

```python
import re
# Define the function to normalize text
def normalize_text(text):
    """text cleaning just for whitespaces and amc sentence markers"""
    # remove <s>
    text = re.sub(r'<s>', '', text)
    text = re.sub(r'</s>', '', text)
    text = text.lower()  # convert to lowercase
    text = text.strip() # remove leading and trailing whitespace
    return text
```

```python
df['text'] = df['Sentence'].apply(normalize_text)
```

```python
df['first_three_words'] = df['text'].str.split().str[:3].str.join(' ')
```

```python
df
```

```python
# Split text into words, get first and last word, and check equality
df['same_first_last'] = df['first_three_words'].apply(lambda x: x.split()[0] == x.split()[-1] if len(x.split()) > 1 else False)

# Show rows where first and last words are the same
matching_rows = df[df['same_first_last']]
print(f"Found {len(matching_rows)} sentences with matching first and last words")
```

```python
df = df[df['same_first_last'] == True] 
```
